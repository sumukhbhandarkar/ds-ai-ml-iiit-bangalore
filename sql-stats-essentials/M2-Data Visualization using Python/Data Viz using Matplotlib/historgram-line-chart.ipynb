```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time series data is commonly uses line charts
months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', "October", "November", "December"])
sales = np.array([5, 85, 150, 120, 130, 100, 120, 187, 199, 214, 200, 255])
```

```python
plt.plot(months, sales)
plt.title("Sales vs Months", fontdict={'fontsize': 20, 'fontweight': 'bold'})
plt.xlabel("Months", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'red'})
plt.ylabel("Sales", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'red'})
ticks = np.arange(0, 250, 25)
labels = ["{}K".format(i//10) for i in ticks]
plt.yticks(ticks, labels)
plt.xticks(rotation=90)

for xy in zip(months, sales):
    plt.annotate(text = "{}".format(xy[1]//10), xy = xy, textcoords="data")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
profit = np.array([-100, 50, 250, -210, -450, 100, 80, 85, -52, 56, -620, 90])
```

```python
plt.hist(profit, bins=25, edgecolor='Orange', color='blue')
plt.show()
```

```python
out1, bins, out3 = plt.hist(profit, bins=50, edgecolor='Orange', color='green')
plt.xticks(bins, rotation=90)
plt.show()
```
