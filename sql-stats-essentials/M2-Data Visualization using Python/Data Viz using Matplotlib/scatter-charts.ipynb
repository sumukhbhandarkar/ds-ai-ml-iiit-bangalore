```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data
sales = np.array([1013.4, 8298.48, 875.51, 22320.83, 9251.6, 4516.5, 585.16, 174.2])
profit = np.array([-1213.46, 1814.13, -1814.13, -2875.32, 985.23, -250.25, 2584.10, -500])
product_category = np.array(['Technology', 'Technology', 'Furniture', 'Technology', 'Office Supplies', 'Technology', 'Technology', 'Office Supplies'])
country = np.array(['Zimbabwe', 'Spain', 'India', 'Pakistan', 'USA', 'UK', 'Australia', 'Georgia'])
```

```python
plt.scatter(profit, sales, alpha=0.7, s=50)
plt.title("Sales vs Profit across countries", fontdict={'fontsize': 20, 'fontweight': '5'})

plt.xlabel("Profit", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})
plt.ylabel("Sales", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})
```

```python
plt.scatter(profit[product_category=="Office Supplies"], sales[product_category=="Office Supplies"], alpha=0.7, s=50, c='Yellow', label='Office Supplies')
plt.scatter(profit[product_category=="Technology"], sales[product_category=="Technology"], alpha=0.7, s=50, c='Blue', label='Technology')
plt.scatter(profit[product_category=="Furniture"], sales[product_category=="Furniture"], alpha=0.7, s=50, c='Cyan', label='Furniture')

plt.legend() #required to show the labels

plt.xlabel("Profit", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})
plt.ylabel("Sales", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})


for xy in zip(profit[country=='India'], sales[country=='India']):
    plt.annotate(text="India", xy=xy)

```
