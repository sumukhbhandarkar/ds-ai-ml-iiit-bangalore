```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_tech = np.array([1013.4, 8298.48, 875.51, 22320.83, 9251.6, 4516.5, 585.16, 174.2])
sales_furniture = np.array([-1213.46, 1814.13, -1814.13, -2875.32, 985.23, -250.25, 2584.10, -500])
sales_office_supplies = np.array([-123.46, 1774.13, -814.13, -275.32, 1985.23, -1250.25, 2584.10, -5000])
```

```python
plt.boxplot([sales_tech, sales_furniture, sales_office_supplies])
plt.title('Sales Boxplot')
plt.xlabel('Product Type', fontdict={'color': 'red'})
plt.ylabel('Sales', fontdict={'color': 'red'})

plt.xticks((1,2,3), ["Tech", "Furniture", "Supplies"])
plt.show()
```

Box plots divide the data range into three important categories, which are as follows:

Median value: This is the value that divides the data range into two equal halves, i.e., the 50th percentile.
Interquartile range (IQR): These data points range between the 25th and 75th percentile values.
Outliers: These are data points that differ significantly from other observations and lie beyond the whiskers.
